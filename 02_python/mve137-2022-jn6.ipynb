# Balls and bins
Verifying exponential approximation and accuracy approximation on the probability that no bin contains more than 1 ball

```python
from math import comb,factorial
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300


j=30

n=np.arange(50,1000)

y1 = 1-j/n

y2 = np.exp(-j/n)
fig, ax = plt.subplots(2)
fig.tight_layout(h_pad=4, w_pad=4)
ax[0].plot(n,y1,'r-',label="$1-j/n$")
ax[0].plot(n,y2,'b-',label="$e^{-j/n}$")
ax[0].legend()
ax[0].grid()
ax[0].set(title='$j=30$',xlabel='$n$')


m=np.arange(1,100)
n=365

p1 = np.zeros(99)

for mm in m:
    
    p1[mm-1]=1
    
    for jj in range(1,mm):
        
        p1[mm-1] = p1[mm-1]*(1-jj/n)


#p1[mm-1] = comb(n,mm)*factorial(mm)/(n**mm)  
    

p2= np.exp(-m**2/(2*n))

ax[1].plot(m,p1,label='exact')

ax[1].plot(m,p2,label='approx.')

ax[1].set(xlabel='$m$',title='$\mathbb{P}[{\overline{\mathcal{E}}}]$, $n=365$')

ax[1].grid()

ax[1].legend()
```
