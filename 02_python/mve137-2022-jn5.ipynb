## Chernoff bounds

We next verify some of the inequalities that we used in the proof of the upper and lower deviation Chernoff bounds for sum of independent Bernoulli random variables  

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300 #to increase resolution figures 

delta = np.arange(0.01,1,0.01)

ud1 = np.exp(delta)/((1+delta)**(1+delta))
ud2 = np.exp(-(delta**2)/3)
ld1 = np.exp(-delta)/((1-delta)**(1-delta))
ld2 = np.exp(-(delta**2)/2)

fig,ax = plt.subplots()

ax.plot(delta,ud1,label='${e^\delta}/{(1+\delta)^{1+\delta}}$')
ax.plot(delta,ud2,label='$e^{-\delta^2/3}$')
ax.set(xlabel='$\delta$')
ax.legend()
ax.set_ylim((0.5,1))
ax.grid()
```

```python
fig,ax = plt.subplots()

ax.plot(delta,ld1,label='$e^{-\delta}/{(1-\delta)^{1-\delta}}$')
ax.plot(delta,ld2,label='$e^{-\delta^2/2}$')
ax.set(xlabel='$\delta$')
ax.legend()
ax.set_ylim((0.5,1))
ax.grid()
```

We now verify that $(e^t + e^{-t})/2\leq e^{t^2/2}$

```python
t=np.arange(-2,2,0.1)
lhs = (np.exp(t)+np.exp(-t))/2
rhs = np.exp(t**2/2)

fig,ax = plt.subplots()

ax.plot(t,lhs,label='$(e^{t}+e^{-t})/2$')
ax.plot(t,rhs,label='$e^{t^2/2}$')
ax.set(xlabel='$t$')
ax.legend()
ax.grid()
```
