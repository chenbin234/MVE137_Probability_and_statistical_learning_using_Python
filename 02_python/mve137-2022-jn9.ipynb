```python
from scipy.stats import bernoulli

import itertools

import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 200


p = 0.7

samples = bernoulli.rvs(p,size=5000)

samplesnorm = 2*samples-1

state = list(itertools.accumulate(samplesnorm))

plt.plot(state)
plt.grid()
plt.xlabel('time')
plt.ylabel('position')
```
